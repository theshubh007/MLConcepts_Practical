 In normalization, we map the minimum feature value to 0 and the maximum to 1. Hence, the feature values are mapped into the [0, 1] range

 In standardization, we donâ€™t enforce the data into a definite range. Instead, we transform to have a mean of 0 and a standard deviation of 1:

```python
import numpy as np
import pandas as pd

df = pd.read_csv("../data/Advertising.csv")
x = df[["TV","radio", "newspaper"]]
y = df["sales"]
```

```python
from sklearn.model_selection import train_test_split

x = np.array(x).reshape(-1,3)
y = np.array(y).reshape(-1,1)

# Split the dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)

x_train[:5]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
print(model.score(x_test, y_test))
print(model.predict([[180, 10, 58]]))

y_predict = model.predict(x_test)
```

```python
import seaborn as sns

sns.displot(y_predict - y_test, kind="kde")
```
